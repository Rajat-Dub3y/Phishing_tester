```python
import pandas as pd

# Load CSV file
df = pd.read_csv("../dataset/phising.csv")  # change filename if needed

# Show first 5 rows
df.head()
```

```python
# Shape of dataset
df.shape

# Column names
df.columns

# Check data types
df.dtypes
```

```python
# Check for null values
df.isnull().sum()
```

```python
# Drop rows with missing values (simple approach)
df = df.dropna()
```

```python
df["Result"].value_counts()
```

```python
# Check duplicate rows
df.duplicated().sum()

# Remove duplicates if any
df = df.drop_duplicates()
```

```python
# Check if any column is non-numeric
non_numeric = df.select_dtypes(exclude=["number"]).columns
non_numeric
```

```python
X = df.drop("Result", axis=1)
y = df["Result"]

print(X.shape, y.shape)
```

```python
# Convert -1 (phishing) to 0, 1 (legit) to 1
y_binary = y.map({-1: 0, 1: 1})

y_binary.value_counts()
```

```python
# Save cleaned dataset
df.to_csv("../dataset/phishing_cleaned.csv", index=False)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y_binary, test_size=0.2, random_state=42, stratify=y_binary
)

X_train.shape, X_test.shape
```
